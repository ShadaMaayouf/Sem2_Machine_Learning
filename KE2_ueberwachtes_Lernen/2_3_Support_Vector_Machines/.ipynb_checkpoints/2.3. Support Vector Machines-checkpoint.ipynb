## 2.3.1. Grundlagen: linear-separierbare Daten <a name="Grundlagen"></a>

- Die **Support Vector Machine (SVM)** ist eine häufig verwendete Methode zur Klassifikation.
- Die Kernidee von SVMs besteht darin, eine Hyperebene im Merkmalsraum zu bestimmen, die die Beispiele der verschiedenen Klassen voneinander trennt.
- Im 2-dimensionalen Raum entsprechen Hyperebenen einfachen Geraden.
- Die Klassifikationsgrenzen des logistischen Modells entsprechen diesen separierenden Hyperebenen.
- Bei der logistischen Regression werden die Klassifikationsgrenzen nur implizit berechnet und der Lernalgorithmus basiert auf der Minimierung der logistischen Kostenfunktion.
- Im Gegensatz dazu wird bei der SVM die Klassifikationsgrenze explizit gelernt.
- Bei binärer Klassifikation wird die Hyperebene gesucht, die die beiden Klassen voneinander trennt und dabei den größtmöglichen Abstand zu den Beispielen der beiden Klassen hat.
- **linear-siparierbare Daten**, d. h., dass es tatsächlich eine Hyperebene gibt, die die Klassen voneinander trennen kann (also kein Beispiel fehlklassifiziert wird).

> Eine Hyperebene in einem n-dimensionalen Raum (Rn) ist ein Unterraum mit einer Dimension weniger als der umgebende Raum. In diesem Fall ist die Hyperebene definiert durch die Gleichung: $$h_{SVM_{\theta,b}} : \theta^T x - b = 0$$
> Hierbei ist: $\theta$ ein Vektor in Rn, $b$ ein Skalar in R, $x$ ein Punkt in Rn, und $\theta^T x$ das Skalarprodukt von $\theta$ und $x$.
> Die Hyperebene besteht aus allen Punkten $x$ in Rn, die diese Gleichung erfüllen.
> Im Kontext von Support Vector Machines (SVMs) repräsentiert diese Hyperebene die Entscheidungsgrenze des Klassifikators. Die Parameter $\theta$ und $b$ werden während des Trainings der SVM gelernt.

#### Beispiel:
Wir erklären hier eine Methode zur Klassifizierung von Äpfeln basierend auf ihrem Durchmesser und Gewicht, die auf Support Vector Machines (SVMs) basiert.

```python
D_fruits = pd.read_csv("./uebung_2_3/sl_svm_ex_fruits.csv")
D_fruits
```

In dem Datensatz `D_fruits` repräsentieren die Spalten `diameter` und `weight` die Merkmale (oder Dimensionen) unserer Datenpunkte, und die Spalte `ok` repräsentiert die Klassenlabels. 

```python
# Trennen Sie die Merkmale und die Zielvariable
X = D_fruits.iloc[:, 0:-1].values
y = D_fruits.iloc[:, 2].values
```



Wir suchen also eine Hyperebene, die durch die Gleichung $$h_{SVM_{\theta,b}} : \theta^T x - b = 0$$ bzw.
$$h_{SVM_{(a_1,a_2),b}} : a_1x_1 + a_2x_2 - b = 0$$ definiert ist, um die Datenpunkte zu trennen. Hierbei ist:
- $\theta$ der Normalenvektor der Hyperebene,
- $x$ ist ein Punkt im Raum (in diesem Fall ein Datenpunkt $(x_1,x_2)$ repräsentiert durch `diameter` und `weight`),
- und $b$ ist der Abstand der Hyperebene vom Ursprung. 

Die Parameter $\theta$ und $b$ werden durch ein Optimierungsverfahren bestimmt, das den Abstand (Margin) zwischen der Hyperebene und den nächstgelegenen Datenpunkten jeder Klasse maximiert. Dies ist das grundlegende Prinzip der SVM-Klassifikation und wird als **Maximierung des Margins** bezeichnet.

```python
def svm_hyperebene_ausgeben(theta, b):
    dimensions = len(theta[0])
    equation = "h(x) : "
    
    for i in range(dimensions):
        equation += str(round(theta[0][i], 2))+"*x_"+str(i+1)
        if i < dimensions - 1:
            equation += " + "
    
    equation += " - "+str(round(b[0], 2))+" = 0"
    print(equation)
    return equation

def svm_optimale_hyperebene_berechnen(theta, b, x_values):
    return (b[0] - theta[0][0] * x_values) / theta[0][1]
    
# Angenommen, du hast bereits die Parameter theta und b berechnet
theta = [[-0.9, -1]]
b = [140]

svm_hyperebene_ausgeben(theta, b)
```

### Schritte zur SVM berechnung:
Wir erstellen wir zuerst eine SVM mit einem linearen Kernel. Dann trainieren wir die SVM mit den Merkmalen ‘diameter’ und ‘weight’ und den Klassenlabels ‘ok’.
Wie erhalten einen Klassifikator `clf`.

```python
# Erstellen Sie das SVM-Modell und Trainieren ihn
clf =  SVC(kernel='linear').fit(X, y)

#Predict the response for test dataset
y_pred = clf.predict(X)

plt.figure()
plt.scatter(D_fruits[y==0]["diameter"], D_fruits[y==0]["weight"], marker='x')
plt.scatter(D_fruits[y==1]["diameter"], D_fruits[y==1]["weight"], marker='s')
plt.show()
```

Nach dem Training können wir die Attribute des trainierten Klassifikators `clf` verwenden, um die Parameter der Hyperebene $θ$ und $b$ zu berechnen. 
- Wir berechnen $θ$ mit `clf.coef_`
- und $b$ mit `clf.intercept_`.

Diese Parameter repräsentieren die Klassifikationsgrenze, die von der SVM gelernt wurde.

```python
# Die Parameter theta und b
theta = clf.coef_
b = clf.intercept_

print("θ ist "+str(theta)+" und b ist "+str(b))
```

Wir definieren dazu zwei Hyperebenen, die durch die folgenden Gleichungen gegeben sind:

- $$h1 = h_{SVM_{(-0.9,-1),-140}} : -0.9x_1 -x_2 +140 = 0$$
- $$h2 = h_{SVM_{(-0.3,-1),-106}} : -0.3x_1 -x_2 +106 = 0$$

```python
x = np.linspace(X[:,0].min(), X[:,0].max(), 100)

# Berechne die zugehörigen y-Werte (weight) für die Hyperebene
h1 = svm_optimale_hyperebene_berechnen([[-0.9, -1]], [-140], x)
h2 = svm_optimale_hyperebene_berechnen([[-0.3, -1]], [-106], x)

# Zeichne die Datenpunkte
plt.figure(figsize=(10,6))
plt.scatter(D_fruits[y==0]["diameter"], D_fruits[y==0]["weight"], marker='x')
plt.scatter(D_fruits[y==1]["diameter"], D_fruits[y==1]["weight"], marker='s')
# Zeichne die Hyperebene
plt.plot(x, h1, '-r')
plt.plot(x, h2, '-b')

plt.xlabel('Diameter')
plt.ylabel('Weight')
plt.title('Hyperebene für SVM-Klassifikation')
plt.show()
```

#### Die lineare Klassifikation

Hier ist die Idee, dass wir eine Hyperebene (definiert durch die Parameter θ und b) haben und wir klassifizieren ein neues Beispiel $x$. Das erfolgt durch die Berechnung, auf welcher Seite der Hyperebene $h_{SVM_{\theta,b}}$ sich $x$ befindet und Ausgeben der entsprechenden Klasse. Die Klassifikationsfunktion $clf_{\theta,b}(x)$ ist also wie folgt definiert:

$$
clf_{\theta,b}(x) = \begin{cases}
   1 &\text{falls } \theta^T x - b ≥ 0 \\
   0 &\text{falls } \theta^T x - b < 0
\end{cases}
$$

Diese Funktion gibt `1` aus (d.h., `clfθ,b(x) = 1`), wenn der Punkt $x$ auf der gleichen Seite der Hyperebene liegt wie die meisten Datenpunkte der Klasse 1 (was wir als "ok" interpretieren könnten), und `0` (d.h., `clfθ,b(x) = 0`), wenn er auf der anderen Seite liegt, die hauptsächlich Datenpunkte der Klasse 0 enthält (was wir als "nicht ok" interpretieren könnten).

chnet.

#### SVM-Klassifikation

In der **binären Klassifikation** mit **Support Vector Machines (SVMs)** haben wir zwei Klassen von Daten, die wir normalerweise als **1** und **-1** bezeichnen. Unser Ziel ist es, eine Hyperebene zu finden, die diese beiden Klassen trennt.

Die Hyperebene wird durch die Parameter **θ** (Theta) und **b** definiert. Die optimalen Werte für diese Parameter finden wir durch die Lösung eines **Minimierungsproblems**, das folgendermaßen aussieht:
$$
\begin{align*}
\min \quad & ||\theta|| \\
\text{s.t.} \quad & \theta^T x - b \geq 1, \quad \forall (x, 1) \in D \\
& \theta^T x - b \leq -1, \quad \forall (x, 0) \in D
\end{align*}
$$

Vereinfacht heißt es $$\begin{align*}
\min \quad & ||\theta|| \\
\text{s.t.} \quad & y(\theta^T x - b) \geq 1, \quad \forall (x, y) \in D
\end{align*}
$$
bei uns im Code dargestellt durch die `svm_optimale_hyperebene_berechnen`-Methode.

Dieses Minimierungsproblem (auch *hard-margin SVM* genannt)besagt im Grunde folgendes:
1. Wir wollen den Betrag von θ minimieren. Das bedeutet, wir suchen nach der kleinstmöglichen Hyperebene, die die Daten noch trennt.

2. Gleichzeitig müssen wir sicherstellen, dass alle Datenpunkte der Klasse 1 auf der einen Seite der Hyperebene liegen und alle Datenpunkte der Klasse -1 auf der anderen Seite. Das drücken wir mathematisch so aus: Für alle Datenpunkte (x,1) muss gelten θ T x - b ≥ 1 und für alle Datenpunkte (x,-1) muss gelten θ T x - b ≤ -1.

Die optimal angepasste Hyperebene (Entscheidungsgrenze) $$h_{0} : \theta^T x - b = 0$$ 
die Grenzen des Margins:
$$h_{1} : \theta^T x - b = 1$$ 
$$h_{-1} : \theta^T x - b = -1$$ 

Mit anderen Worten, wir suchen nach der schmalsten möglichen "Straße", die zwischen den Datenpunkten der beiden Klassen verläuft, und die Hyperebene ist die Mittellinie dieser Straße. Die Breite der Straße ist dabei immer mindestens 2 (da wir auf der einen Seite ≥ 1 und auf der anderen Seite ≤ -1 haben), und wir versuchen, sie so breit wie möglich zu machen, indem wir θ minimieren.
Mit anderen Worten, h1 und h−1 sind die beiden im Abstand von jeweils $\frac{1}{||\theta||}$ liegenden Hyperebenen zu $h_0$.

> - In der Praxis benennt man die beiden Klassen in der binären Klassifikation mit SVMs üblicherweise als 1 und -1 anstelle von 1 und 0, um die mathematische Behandlung zu vereinfachen.

Die Methode `DecisionBoundaryDisplay.from_estimator()` aus der `scikit-learn` Bibliothek kann verwendet werden, um die Entscheidungsgrenze eines Klassifikators zu visualisieren. Die Parameter `plot_method`und `levels` sorgen dafür, dass die Entscheidungsgrenze und die Margins gezeichnet werden.
> - `levels= 0` --> Ein Margen wird gezeichnet.
> - `levels=[-1,0,1]` --> Alle Margen werden gezeichnet.

```python
plt.figure()
DecisionBoundaryDisplay.from_estimator(clf,X,plot_method="contour",levels=[-1,0,1])

plt.scatter(D_fruits[y==0]["diameter"], D_fruits[y==0]["weight"], marker='x')
plt.scatter(D_fruits[y==1]["diameter"], D_fruits[y==1]["weight"], marker='s')
plt.show()
```

> **Unterschied zwischen der linearen und der SVM-Klassifikation**
>
> Der Hauptunterschied besteht darin, dass die SVM versucht, den Abstand zwischen den Klassen zu maximieren (d.h., sie sucht nach der breitesten möglichen “Straße” zwischen den Klassen), während die lineare Klassifikation einfach eine Hyperebene zieht, die die Klassen trennt, ohne notwendigerweise den Abstand zu maximieren.



```python
print("Die optimale Hyperebene ist definiert durch "+svm_hyperebene_ausgeben(theta, b))
```

```python
h_opt = svm_optimale_hyperebene_berechnen(theta, b, x)

plt.figure()
DecisionBoundaryDisplay.from_estimator(clf,X,plot_method="contour",levels=0)
#plt.plot(x, h_opt, '-r')
plt.scatter(D_fruits[y==0]["diameter"], D_fruits[y==0]["weight"], marker='x')
plt.scatter(D_fruits[y==1]["diameter"], D_fruits[y==1]["weight"], marker='s')
plt.show()
```

```python
print("Accuracy:", accuracy_score(y, y_pred))
# Model Precision: what percentage of positive tuples are labeled as such?
print("Precision:", precision_score(y, y_pred, zero_division=0))

# Model Recall: what percentage of positive tuples are labelled as such?
print("Recall:", recall_score(y, y_pred, zero_division=0))
```

## 2.3.2. Nicht-linear-separierbare Daten <a name="nl_daten"></a>

In realen Anwendungsfällen sind Daten oft nicht linear trennbar, was bedeutet, dass es keine gerade Linie (oder in höheren Dimensionen eine Hyperebene) gibt, die die Klassen perfekt trennt (Daten sind nicht linear trennbar). In solchen Fällen kann das Standard-Optimierungsproblem der Support Vector Machine (SVM) keine Lösung finden.

Um dieses Problem zu lösen verwendet die SVM eine **Hinge-Kostenfunktion** und einen **Regularisierungsparameter**, um das Optimierungsproblem zu modifizieren und trotzdem eine möglichst gute Trennung der Daten zu erreichen.
Anstatt strikt zu fordern, dass alle Datenpunkte korrekt klassifiziert werden, wird die Hinge-Kostenfunktion benutzt, um zu missen, wie gut ein Datenpunkt klassifiziert wird, und bestraft Fehlklassifikationen.

Die Hinge-Kostenfunktion ist definiert wie folgt: $$L^{\text{hinge}}(D, f) = \sum_{i=1}^{m} \max\{0, 1 - y^{(i)}f(x^{(i)})\}$$

Wenn ein Datenpunkt perfekt klassifiziert wird, trägt er nichts zu den Kosten bei. Wenn er falsch klassifiziert wird, erhöhen die Kosten. Dieses modifizierte Optimierungsproblem enthält auch einen Regularisierungsparameter $C$, der kontrolliert, wie stark Fehlklassifikationen bestraft werden. Die optimalen Parameter für das Modell werden dann durch die Lösung dieses Optimierungsproblems bestimmt.

Für einen Datensatz D sind damit die optimalen Parameter θ und b bestimmt durch die Lösung des folgenden Minimierungsproblems (auch **soft-margin SVM'** genannt):
$$\min C ||\theta||^2 + \frac{1}{m} \sum_{i=1}^{m} \max\{0, 1 - y^{(i)}(\theta^T x^{(i)} - b)\}$$


```python
df = pd.read_csv("./uebung_2_3/sl_logreg_ex_apartment.csv")

X = df.iloc[:, 0:-1].values
y = df.iloc[:, 2].values

clf = SVC(kernel='linear', C=1).fit(X, y)

plt.figure()
DecisionBoundaryDisplay.from_estimator(clf,X,plot_method="contour",levels=0)

plt.scatter(df[y==0]["size"], df[y==0]["dist"], marker='x')
plt.scatter(df[y==1]["size"], df[y==1]["dist"], marker='s')
plt.show()
```

## 2.3.3. Kernelfunktionen <a name="kernelfunktionen"></a>
Wir betrachten den Datensatz $D_{screws}$. Es sollte offensichtlich sein, dass weder eine lineare hard-margin 
noch eine lineare soft-margin SVM die verschiedenen Klassen erkennen kann
Bei SVMs können wir jedoch einen Trick anwenden, den sogenannten **Kernel-Trick**, der uns eine explizite Merkmalsausweitung erspart und somit den rechnerischen Aufwand reduziert. 
:<ins> wir transformieren den Ursprungs-Merkmalsrau 
in einenöo¨herdimensionalen Raum, indem wir neue Merkmale durch Kombination existierender Merkmale erzeug</ins>.en.

```python
df = pd.read_csv("./uebung_2_3/sl_logreg_ex_screws.csv")

X = df.iloc[:, 0:-1].values
y = df.iloc[:, 2].values

plt.figure()
plt.scatter(df[y==0]["weight"], df[y==0]["length"], marker='x')
plt.scatter(df[y==1]["weight"], df[y==1]["length"], marker='s')
plt.show()
```

Es gibt verschiedene Arten von Kernel-Funktionen, wie z.B. den *linearen Kernel* `kernel='linear'`, den *Polynom-Kernel* `kernel='poly'` und den *Radial Basis Function (RBF) Kernel* `kernel='rbf'`. Jeder dieser Kernel hat seine eigenen Eigenschaften und ist für verschiedene Arten von Daten geeignet. Hier verwenden wir den **Radial Basis Function (RBF) Kernel**.
> - `C=1` ist der Regularisierungsparameter. Ein kleinerer Wert von C erzeugt eine breitere “Straße” aber möglicherweise mehr Fehlklassifikationen, während ein größerer Wert von C eine engere “Straße” erzeugt und versucht, alle Trainingsexemplare korrekt zu klassifizieren.
> - gamma=1 ist ein Parameter für den RBF-Kernel. Ein kleinerer Wert von gamma wird dazu führen, dass jeder einzelne Trainingseintrag einen weitreichenden Einfluss hat, während ein größerer Wert von gamma bedeutet, dass jeder Trainingseintrag nur einen begrenzten Einflussbereich hat.


Wir zeigen die Entscheidungsgrenze zur SVM basierend auf der Radialen Basisfunktion mit Parametern $γ=1$ und $C=500$. Punkte innerhalb der Grenze werden als ”ok“ klassifiziert und Punkte außerhalb der Grenze werden als ”nicht ok“ klassifiziert.


```python
clf = SVC(kernel='rbf',C=500,gamma=1).fit(X, y)

plt.figure()
DecisionBoundaryDisplay.from_estimator(clf,X,plot_method="contour",levels=[0])
plt.scatter(df[y==0]["weight"], df[y==0]["length"], marker='x')
plt.scatter(df[y==1]["weight"], df[y==1]["length"], marker='s')
plt.show()
```

### lineares SVM:  <a name="zf_l_svm"></a>

```python
def svm_hyperebene_ausgeben(theta, b):
    dimensions = len(theta[0])
    equation = "h(x) : "
    
    for i in range(dimensions):
        equation += str(round(theta[0][i], 2))+"*x_"+str(i+1)
        if i < dimensions - 1:
            equation += " + "
    
    equation += " - "+str(round(b[0], 2))+" = 0"
    print(equation)
    return equation

def svm_optimale_hyperebene_berechnen(theta, b, x_values):
    return (b[0] - theta[0][0] * x_values) / theta[0][1]
    
```

```python
D_fruits = pd.read_csv("./uebung_2_3/sl_svm_ex_fruits.csv")
# Trennen Sie die Merkmale und die Zielvariable
X = D_fruits.iloc[:, 0:-1].values
y = D_fruits.iloc[:, 2].values

# Erstellen Sie das SVM-Modell und Trainieren ihn
clf =  SVC(kernel='linear').fit(X, y)

#Predict the response for test dataset
y_pred = clf.predict(X)

# Die Parameter theta und b
theta = clf.coef_
b = clf.intercept_

print("θ ist "+str(theta)+" und b ist "+str(b))

x = np.linspace(X[:,0].min(), X[:,0].max(), 100)

print("Die optimale Hyperebene ist definiert durch "+svm_hyperebene_ausgeben(theta, b))
# Berechne die zugehörigen y-Werte (weight) für die Hyperebene
h1 = svm_optimale_hyperebene_berechnen([[-0.9, -1]], [-140], x)
h2 = svm_optimale_hyperebene_berechnen([[-0.3, -1]], [-106], x)

# Zeichne die Datenpunkte
plt.figure(figsize=(10,6))
plt.scatter(D_fruits[y==0]["diameter"], D_fruits[y==0]["weight"], marker='x')
plt.scatter(D_fruits[y==1]["diameter"], D_fruits[y==1]["weight"], marker='s')
# Zeichne die Hyperebene
plt.plot(x, h1, '-r')
plt.plot(x, h2, '-b')

plt.xlabel('Diameter')
plt.ylabel('Weight')
plt.title('Hyperebene für SVM-Klassifikation')
plt.show()
```

```python
# Zeichne die Datenpunkte
plt.figure(figsize=(10,6))
DecisionBoundaryDisplay.from_estimator(clf,X,plot_method="contour",levels=[-1,0,1])

plt.scatter(D_fruits[y==0]["diameter"], D_fruits[y==0]["weight"], marker='x')
plt.scatter(D_fruits[y==1]["diameter"], D_fruits[y==1]["weight"], marker='s')

plt.xlabel('Diameter')
plt.ylabel('Weight')
plt.title('Hyperebene für SVM-Klassifikation')
plt.show()
```
