In vorangegangenen Abschnitten wurden Unabhängigkeitsannahmen zu Merkmalen getroffen, um bestimmte Gleichungen zu rechtfertigen, wie beispielsweise bei der naiven Bayes-Klassifikation. In der Praxis treten jedoch oft Merkmale mit korrelierenden Ausprägungen auf, die sich negativ auf den Lernprozess auswirken können.
- Ein triviales Beispiel dafür ist die Dopplung eines Merkmals, wie etwa die Größe einer Person in cm und mm. Das zweite Merkmal bringt keine zusätzliche Information und kann zu einer unbeabsichtigten Gewichtung im gelernten Modell führen. Durch das Weglassen eines dieser Merkmale kann der Datensatz kompakter dargestellt werden, ohne Information zu verlieren. Dies kann insbesondere bei großen Datensätzen und komplexen Lernalgorithmen zur Verringerung der Laufzeit beitragen.
- Ähnliche Effekte treten auch bei Merkmalen auf, die nicht perfekt korrelieren, wie beispielsweise Größe und Gewicht einer Person.
- Es kann auch vorkommen, dass ein Merkmal als eine ungefähre Kombination von mehr als einem anderen Merkmal dargestellt werden kann. In solchen Fällen kann es sinnvoll sein, die Daten mit weniger Merkmalen zu beschreiben.

>**!!**
>
>Es ist nicht immer einfach zu erkennen, wie sich Merkmale zueinander verhalten.

## 3.5.1. Grundlagen <a name="3_5_1"></a>

Die **Hauptkomponentenanalyse (PCA)** ist eine weit verbreitete Methode zur Lösung von Problemen, die mit korrelierenden Merkmalen zusammenhängen. Hier sind die Hauptpunkte:

- PCA ist eine Methode zur **Dimensionsreduktion**. Sie transformiert einen Datensatz in $\mathbb{R}^n$ in einen Datensatz in $\mathbb{R}^{n'}$ mit $n' < n$, wobei möglichst wenig Information verloren geht.
- PCA kann als Vorverarbeitungsschritt für überwachtes oder unüberwachtes Lernen verwendet werden, oder einfach nur um einen Datensatz zu komprimieren.
- PCA selbst kann auch <ins>als eine Methode des unüberwachten Lernens</ins> verstanden werden, da sie nach Mustern im Datensatz sucht.
- Die allgemeine Methodik der PCA ist mathematisch recht komplex. Zunächst wird die Intuition der Methode bei der Reduktion eines Datensatzes vom $\mathbb{R}^2$ zum $\mathbb{R}^1$ diskutiert, und erst in [Abschnitt 3.5.3](#3_5_3) wird die allgemeine Methode eingeführt.

#### Standardisierung
Um PCA anwenden zu können, ist die **Standardisierung** der Daten <ins>notwendig</ins>. Durch die Standardisierung wird sichergestellt, dass jedes Merkmal den gleichen Einfluss auf die PCA hat, unabhängig von seiner Skala oder den Einheiten, in denen es gemessen wird.

Die `z-Transformation` ist eine gängige Methode zur Standardisierung von Daten ([siehe auch Abschnitt 2.4.2](http://127.0.0.1:8888/lab/tree/KE2_ueberwachtes_Lernen/2_4_Nachste_Nachbarn_Klassifikation/2_4_Naechste_Nachbarn_Klassifikation.ipynb)). Sie transformiert die Daten so, dass sie **einen Mittelwert von 0** und eine **Standardabweichung von 1** haben. 

Hier ist ein Beispiel, wie Sie die z-Transformation mit der `StandardScaler`-Funktion aus der `sklearn.preprocessing`-Bibliothek in Python durchführen können:
```python
from sklearn.preprocessing import StandardScaler

# Angenommen, X ist Ihr Datensatz
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)
```
Nachdem die Daten standardisiert sind, können Sie die PCA anwenden.

Die Hauptkomponentenanalyse (PCA) wird verwendet, um die Dimensionalität eines Datensatzes zu reduzieren, indem sie einen Unterraum findet, der die Varianz der Daten <ins>maximiert</ins>. Bei der Anwendung von PCA auf ein überwachtes Lernproblem können wir die Klassen der Beispiele ignorieren, da die Dimensionsreduktion nur auf den Datenpunkten basiert.

Angenommen, wir haben einen Datensatz $E = \{x^{(1)}, ..., x^{(m)}\}$ mit $x^{(i)} \in \mathbb{R}^2$, $i = 1,...,m$. PCA versucht, einen 1-dimensionalen Unterraum (eine Gerade) in $\mathbb{R}^2$ zu finden, sodass die quadrierten Distanzen aller Punkte in E zu ihren jeweiligen Projektionen auf diesen Unterraum minimal sind.

Formal löst PCA das folgende Optimierungsproblem:

$$v^* = \arg\min_{v \in \mathbb{R}^2, ||v|| = 1} \sum_{x \in E} ||x - (x^T v)v||^2$$

Der Vektor $v^*$ charakterisiert den gesuchten 1-dimensionalen Unterraum als alle Vielfachen von $v^*$, d.h., die gesuchte Gerade besteht aus allen Punkten $w \in \mathbb{R}^2$ mit $w = xv^*$ für $x \in \mathbb{R}$. 

Der Ausdruck $||x - (x^T v)v||$ ist der Abstand des Punktes x zu der von v aufgespannten Geraden. 

Die zusätzliche Bedingung $||v|| = 1$ fordert, dass v genau Länge 1 hat. Diese Bedingung dient nur der Normierung. Das Problem ist im Allgemeinen  <ins>nicht</ins> eindeutig lösbar, in diesem Fall ist $v^*$ eine beliebige minimale Lösung. Der Vektor $v^*$ wird dann **als (erste) Hauptkomponente von E** bezeichnet.

#### **Beispiel 3**: (Fortsetzung von Beispiel 2)
Abbildung 3 zeigt die (in diesem Fall eindeutig bestimmte) Gerade, die durch den Nullpunkt geht und minimalen Abstand zu allen Punkten aus $\hat{D}_{\text{ships}}$ hat. Insbesondere ist hier $\vec{v}^*_{\text{ships}} \approx (0.707, 0.707)^T$ eine Lösung des Optimierungsproblems (1), also eine erste Hauptkomponente von $\hat{D}_{\text{ships}}$. Abbildung 3 zeigt weiterhin auch die Projektionen der Punkte aus $\hat{D}_{\text{ships}}$ auf diese Gerade an (diese sind allerdings ohne Klassen aufgeführt).


````python
D_ships_scaled = pd.DataFrame(X_scaled, columns=['Länge', 'Verdrängung'])
D_ships_scaled['Typ'] = y
````

````python
# Streudiagramm für jeden Schiffstyp erstellen und labeln
for typ, color in colors.items():
    subset = D_ships_scaled[D_ships_scaled['Typ'] == typ]
    plt.scatter(subset['Länge'], subset['Verdrängung'], c=color, label=labels[typ], marker=markers[typ])

# Gerade zeichnen
x_line = np.linspace(D_ships_scaled['Länge'].min(), D_ships_scaled['Länge'].max()+0.5, 100)
y_line = x_line  # Ändern Sie diese Zeile, um die Steigung und den y-Achsenabschnitt der Geraden anzupassen
plt.plot(x_line, y_line, 'k-', label='Gerade')

# Projektionen der Punkte auf die Gerade zeichnen
for i in range(len(D_ships_scaled)):
    x_point = D_ships_scaled.loc[i, 'Länge']
    y_point = D_ships_scaled.loc[i, 'Verdrängung']
    x_proj = (x_point + y_point) / 2  # Ändern Sie diese Zeile, um die Projektion zu berechnen
    y_proj = x_proj  # Ändern Sie diese Zeile, um die Projektion zu berechnen
    plt.plot([x_point, x_proj], [y_point, y_proj], 'k--')
    plt.scatter(x_proj, y_proj, c='black')

plt.xlabel('Länge (in m) [skaliert]')
plt.ylabel('Verdrängung (in t) [skaliert]')
plt.title('Schiffe: Länge vs Verdrängung (skaliert)')
plt.legend()

plt.show()
````

Der folgebnde Code führt eine Hauptkomponentenanalyse (PCA) auf dem skalierten Datensatz `X_scaled` durch und reduziert seine Dimensionalität auf eine Komponente. 

Die Methode `explained_variance_ratio_` gibt den Anteil der Gesamtvarianz im Datensatz zurück, der von jeder Hauptkomponente erklärt wird. Da wie `n_components=1` gesetzt haben, wird `explained_variance_ratio_` ein Array mit einem einzigen Wert zurückgeben, der den Anteil der Varianz angibt, der von der ersten (und einzigen) Hauptkomponente erklärt wird.

Wenn Wir den Wert von `explained_variance_ratio_` ausdrucken möchten, können wir folgenden Code verwenden:

```python
print(pca.explained_variance_ratio_)
```

Dies wird den Anteil der Gesamtvarianz im Datensatz ausgeben, der von der ersten Hauptkomponente erklärt wird. Ein höherer Wert bedeutet, dass die Hauptkomponente einen größeren Teil der Struktur im Datensatz erklärt. Ein Wert nahe 1 würde bedeuten, dass fast alle Informationen im Datensatz in dieser einen Komponente enthalten sind. Ein Wert nahe 0 würde bedeuten, dass die Komponente sehr wenig von der Struktur im Datensatz erklärt. 

````python
from sklearn.decomposition import PCA

pca = PCA(n_components=1)
pca.fit(X_scaled)

pca.explained_variance_ratio_
````

````python
X_scaled
````

Sobald der Vektor $v^*$ bestimmt ist, können Wir den Datensatz $E$ komprimieren, indem wir für jeden Datenpunkt $x \in E$ anstelle der beiden Koordinaten $x = (x_1, x_2)^T \in \mathbb{R}^2$ nur die Koordinate $z \in \mathbb{R}$ speichern. Hierbei ist $zv^*$ die Projektion von $x$ auf die durch $v^*$ aufgespannte Gerade. 

````python
X_scaled_compressed = pca.transform(X_scaled)
X_scaled_compressed
````

````python
#in dataframe umwandeln
D_ships_compressed = pd.DataFrame(X_scaled_compressed, columns=['z'])
````

In diesem Code ist `X_scaled_compressed` der komprimierte Datensatz. Jede Zeile in `X_scaled_compressed` entspricht einem Datenpunkt in `X_scaled`, und jede Spalte entspricht einer Hauptkomponente. Da wir `n_components=1` gesetzt haben, wird `X_scaled_compressed` nur eine Spalte haben. Diese Spalte enthält die Koordinaten $z$, die wir erwähnt haben.

Nun kann die Ausgabe der PCA, d.h., der komprimierte Datensatz $\hat{D}_{\text{compressed}}^{ships}$für die eigentliche Lernaufgabe verwendet werden, d.h., beispielsweise logistische Regression anwenden, um einen Klassifikator zu erlernen.

Das daraus entstehende Modell (also beispielsweise ein Klassifikator `clf`) ist damit allerdings nur auf 1-dimensionalen Datenpunkten definiert (also `clf` : $\mathbb{R}^1 \rightarrow C$, wobei $C$ die Menge der Klassen ist).

````python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression().fit(X_scaled_compressed,y)
````

Diese ist gegeben durch $z = x^T v^*$ und damit können wir via `clf(x^T v^*)` auch Datenpunkte $x \in \mathbb{R}^2$ des ursprünglichen Problems klassifizieren.

## 3.5.2. Hauptkomponentenanalyse und lineare Regression <a name="3_5_2"></a>

Abbildung 3 zeigt Ähnlichkeiten mit der Methode der linearen Regression (siehe Unterkapitel 2.1) auf und könnte dazu verleiten, die Hauptkomponentenanalyse (PCA) mit der linearen Regression gleichzusetzen. Bei der linearen Regression versucht man, gegeben eine Menge von Datenpunkten im $$\mathbb{R}^2$$ (oder präziser, eine Menge von Beispielen $(x, y)$, sodass $x, y \in \mathbb{R}$ und $x$ ist das einzige Merkmal und $y$ der Funktionswert), eine lineare Funktion `regr` (eine Gerade) zu finden, sodass die Summe der quadrierten Abweichungen der vorhergesagten Funktionswerte `regr(x)` von $y$ minimal ist. Einfach ausgedrückt, minimiert die lineare Regression die Summe der quadrierten "vertikalen Abstände" zu der gesuchten Geraden.

Im Gegensatz dazu minimiert die PCA die tatsächlichen (quadrierten) Abstände zu der Geraden, d.h., die kürzeste Verbindung zur Geraden (beachten Sie, dass diese Anschauung nur im 2-dimensionalen gültig ist). Abbildung 5 veranschaulicht die unterschiedlichen Optimierungsprobleme von PCA und linearer Regression. Für typische 2-dimensionale Probleme sind die Lösungen dieser beiden Optimierungsprobleme nicht identisch, oft aber auch recht ähnlich.

Dennoch handelt es sich bei PCA um eine grundsätzlich andere Methode (was spätestens auch im nächsten Abschnitt 3.5.3 deutlich werden sollte) für einen anderen Zweck als die lineare Regression. Während es bei der linearen Regression um Funktionsapproximation geht, geht es bei PCA um Dimensionsreduktion.

## 3.5.3. Die allgemeine Hauptkomponentenanalyse <a name="3_5_3"></a>

In der allgemeinen Hauptkomponentenanalyse betrachten wir das Problem, einen Datensatz $E = \{x^{(1)}, ..., x^{(m)}\}$ mit $x^{(i)} \in \mathbb{R}^n$ auf einen Datensatz $E_{\text{compressed}} = \{z^{(1)}, ...,z^{(m)}\}$ mit $z^{(i)} \in \mathbb{R}^{n'}$ zu komprimieren, für $i = 1,...,m$ und $n' < n$. Wir gehen davon aus, dass der Datensatz $E$ standardisiert ist, d.h., dass der Mittelwert aller Merkmalsausprägungen 0 ist. 

In diesem allgemeinen Fall suchen wir also einen $n'$-dimensionalen Unterraum von $\mathbb{R}^n$, sodass die Summe der quadrierten Abstände von jedem Punkt $x \in E$ zu seiner Projektion $\hat{x}$ auf diesem Unterraum minimal ist. Einen $n'$-dimensionalen Unterraum kann man charakterisieren durch $n'$ paarweise orthogonale Einheitsvektoren $v^*_1,..., v^*_{n'} \in \mathbb{R}^n$ (diese bilden dann eine orthonormale Basis, kurz ONB, dieses Unterraums). 

Das zugehörige Optimierungsproblem ist damit gegeben durch

$$(v^*_1,..., v^*_{n'}) = \arg\min_{v_1,...,v_{n'} \text{ ist ONB}} \sum_{x \in E} ||x - (x^T v_1)v_1 - ... - (x^T v_{n'})v_{n'}||^2$$

wobei "v_1,..., v_{n'} ist ONB" heißt, dass $||v_1|| = ... = ||v_{n'}|| = 1$ und $v_i^T v_j = 0$ für alle $i, j = 1,...,n'$ mit $i \neq j$. Wie man leicht einsehen sollte, ist das Problem (1) ein Spezialfall des Problems (2). Ist $(v^*_1,..., v^*_{n'})$ eine Lösung von (2), so ist der gesuchte $n'$-dimensionale Unterraum durch die Menge aller Linearkombinationen von $v^*_1,..., v^*_{n'}$ gegeben. 

Die neuen Koordinaten eines Punktes $x \in \mathbb{R}^n$ sind dann definiert durch $z \in \mathbb{R}^{n'}$ mit

$$z = (x^T v^*_1,..., x^T v^*_{n'})$$

Die Vektoren $v^*_1,..., v^*_{n'}$ heißen dann die ersten $n'$ Hauptkomponenten von $E$. Eine Berechnung von $(v^*_1,..., v^*_{n'})$ durch explizite Lösung des Optimierungsproblems (2) ist recht aufwändig und in der Praxis in dieser Form unüblich. Es existiert allerdings eine recht elegante algebraische Charakterisierung der Hauptkomponenten durch die Eigenvektoren der Kovarianzmatrix.

**Definition 1.** Sei $E = \{x^{(1)}, ..., x^{(m)}\}$ ein standardisierter Datensatz. Die Kovarianzmatrix $$\text{Cov}(E) \in \mathbb{R}^{n \times n}$ von $E$ ist definiert durch

$$\text{Cov}(E) = \frac{1}{m} \sum_{i=1}^{m} x^{(i)} (x^{(i)})^T$$

Die Kovarianzmatrix $\text{Cov}(E)$ ist stets symmetrisch und positiv definit, d.h., es existieren genau $n$ reale Eigenwerte $\lambda_1,...,\lambda_n$ (potentiell treten Eigenwerte mehrfach auf) von $\text{Cov}(E)$ mit

$$\lambda_1 \leq \lambda_2 \leq ... \leq \lambda_n$$

Seien nun $w_1,...,w_n$ passende Eigenvektoren mit $||w_1|| = ... = ||w_n|| = 1$, d.h.,

$$\text{Cov}(E)w_i = \lambda_i w_i$$

für alle $i = 1,...,n$. Dann lässt sich zeigen, dass genau diese Eigenvektoren die Hauptkomponenten von $E$ sind:

**Theorem 1.** $(w_1,...,w_{n'})$ ist für jedes $n' < n$ Lösung von (2).

Das obige Theorem (das wir hier nicht beweisen werden) charakterisiert somit die Hauptkomponenten von $E$ als die (zu den zugehörigen Eigenwerten aufsteigend sortierten) Eigenvektoren von $\text{Cov}(E)$.

rinzipiell ohne Bedeutung ist.

## 3.5.4. Evaluation <a name="3_5_4"></a>

Betrachten wir noch einmal Abbildung 3 aus Beispiel 3. Es sollte offensichtlich sein, dass wir bei der Projektion der Datenpunkte auf die erste Hauptkomponente keine "relevanten" Informationen verlieren. Das bedeutet, dass für die anschließende Lernaufgabe alle notwendigen Informationen für die Klassifikation von Datenpunkten auf dem projizierten Unterraum noch vorhanden sind. Beispielsweise sind, bei Verwendung von SVMs, die beiden Klassen sowohl im Ursprungsraum als auch im projizierten Raum linear separierbar.

Die Frage, die man sich bei der Verwendung von PCA stellen sollte, ist, wie weit man die Dimensionalität des Ursprungsraumes reduzieren kann, ohne für die eigentliche Lernaufgabe relevante Informationen zu verlieren (oder ob man überhaupt die Dimensionalität reduzieren sollte). Die direkteste Methode, diese Frage zu beantworten, ist es, die Evaluationsmaße der eigentlichen Lernaufgabe für verschiedene Varianten der Dimensionsreduktion zu Rate zu ziehen. Ist beispielsweise die eigentliche Lernaufgabe die Klassifikation, so können wir (beispielsweise) das F1-Maß für Modelle verschiedener Dimensionalität berechnen und anschließend die Dimensionalität wählen, bei der das F1-Maß maximal bzw. noch "ausreichend hoch" ist. Dieser Ansatz ist jedoch mit einem erheblichen Aufwand verbunden, da wir zum einen Datensätze verschiedener Dimensionalität generieren müssen und zum anderen für jeden dieser Datensätze ein Modell lernen und evaluieren müssen.

Die übliche Methode, um zu bestimmen, auf welche Dimension $n'$ wir einen Datensatz der Dimensionalität $n$ ($n > n'$) komprimieren sollten, ist es, den Verlust an Varianz bei einer Dimensionsreduktion zu messen und anschließend die Dimensionalität wählen, bei der der Verlust an Varianz "nicht zu hoch" ist (hier sparen wir uns also das erneute Lernen eines Modells). Intuitiv beschreibt die Varianz eines Datensatzes $E$ wie weit "verstreut" die Datenpunkte $E$ im Raum liegen. Beispielsweise sehen wir in Abbildung 3, dass die paarweisen Abstände der Punkte zueinander nicht signifikant kleiner unter der Projektion werden, der Verlust an Varianz ist hier also sehr gering. Für einen standardisierten Datensatz $E = \{x^{(1)}, ..., x^{(m)}\}$ ist die Varianz $\text{var}(E)$ definiert als

$$\text{var}(E) = \frac{1}{m} \sum_{i=1}^{m} ||x||^2$$

und ist $E_{\text{compressed}}$ eine komprimierte Variante von $E$, so ist

$$\text{retVar}(E_{\text{compressed}},E) = \frac{\text{var}(E_{\text{compressed}})}{\text{var}(E)}$$

der Anteil der Varianz von $E$, der in $E_{\text{compressed}}$ noch erhalten bleibt (beachten Sie, dass stets $\text{var}(E_{\text{compressed}}) \leq \text{var}(E)$ gilt; $\text{retVar}$ steht hier für retained variance). Üblicherweise wählt man $n'$ so klein wie möglich, sodass

$$\text{retVar}(E_{\text{compressed}},E) \geq \alpha$$ 

für einen Schwellwert $\alpha \in [0,1]$ gilt. Typische Werte für $\alpha$ liegen dabei im Bereich [0.9,0.99].

Beispiel 5. Wir setzen Beispiel 4 fort. Die Varianz $\text{var}(D_{\text{ships}})$ des Datensatzes $D_{\text{ships}}$ (siehe Tabelle 2) berechnet sich zu

$$\text{var}(D_{\text{ships}}) = \frac{1}{6} (||(0.766,0.425)^T||^2 + ||(-0.266,-0.38)^T||^2 + ||(-0.828,-0.908)^T||^2 + ||(1.211,1.624)^T||^2 + ||(0.742,0.601)^T||^2 + ||(-1.625,-1.361)^T||^2) \approx 2$$

Die Varianz $\text{var}(D_{\text{compressed ships}})$ des Datensatzes $D_{\text{compressed ships}}$ (siehe Beispiel 4) berechnet sich zu

$$\text{var}(D_{\text{compressed ships}}) = \frac{1}{6} (||(0.841)||^2 + ||(-0.457)||^2 + ||(-1.228)||^2 + ||(2.004)||^2 + ||(0.95)||^2 + ||(-2.112)||^2) \approx 1.967$$

und somit ist

$\text{retVar}(D_{\text{compressed ships}},D_{\text{ships}}) = \frac{1.967}{2} \approx 0.984$

die bei der Komprimierung erhaltene Varianz (es werden also ungefähr 98.4% der Varianz erhalten). Dieser Wert verdeutlicht, dass der Verlust an Varianz hier prinzipiell ohne Bedeutung ist.
