# K-Mean_Algorithmus

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

data = {
    'X': [2, 2, 5, 4, 8, 6, 5, 5, 6, 4],
    'Y': [4, 6, 6, 7, 3, 6, 2, 7, 3, 4]
}

df = pd.DataFrame(data)

print(df)
```

```python
X = df.values
plt.figure()
plt.scatter(X.T[0], X.T[1])
plt.show()
```


1. Zuerst werden Zentroiden der Anzahl k in der Regel zufällig ausgewählt und in das multidimensionale Koordinatensystem gesetzt.
2. Dann werden die Abstände jeder Beobachtung zu jedem Zentroid berechnet.
3. Die Beobachtungen werden dem Zentroid (Cluster) zugeordnet, dem sie am nächsten sind.
4. Die Zentroide werden in der Mitte des Clusters platziert.
5. Die Schritte 1 bis 4 werden wiederholt, bis es keine weitere Neuzuordnung von Beobachtungen zu Clustern gibt.

Es ist wichtig zu beachten, dass die Anzahl der Cluster k und damit die anfänglichen Zentroide a priori definiert werden müssen. Eine Methode zur Bestimmung der optimalen Anzahl von Clustern könnte das Ellenbogendiagramm sein, das die Varianz gegen die Anzahl der Cluster abbildet¹.

Bitte beachten Sie, dass die Wahl der Anfangszentren und die Anzahl der Cluster einen erheblichen Einfluss auf die Ergebnisse des K-Means-Algorithmus haben können. Daher ist es oft sinnvoll, den Algorithmus mehrmals mit unterschiedlichen Anfangszentren auszuführen und die Ergebnisse zu vergleichen.

### **Aufgabe**: Daten in 3 Cluster aufteilen.

#### 1. Wir wählen die k zufälligen Zentroiden und setzen diese in das multidimensionale Koordinatensystem: 
$ k = 3 $ von daher wählen wir 3 zufällige Zentroiden-Punkte.
$$m_1 = (1, 5)$$
$$m_2 = (4, 1)$$
$$m_3 = (8, 4)$$

```python
m1 = (1, 5)
m2 = (4, 1)
m3 = (8, 4)
      
plt.figure()
plt.scatter(X.T[0], X.T[1], color='black')

'''
In Python, der Asterisk (`*`) wird als Entpackungsoperator verwendet¹. 
Wenn er vor einer Sammlung (wie einer Liste oder einem Tupel) in einem Funktionsaufruf platziert wird, entpackt er die Sammlung und übergibt ihre Elemente als separate Positionalargumente an die Funktion¹.

In Ihrem speziellen Fall ist `m1` beispielsweise `(1, 5)`, dann ist `plt.scatter(*m1, color='red', marker='s')` äquivalent zu `plt.scatter(1, 5, color='red', marker='s')`.

'''

plt.scatter(*m1, color='red', marker = 's')  # m1 wird in Rot dargestellt
plt.scatter(*m2, color='green', marker = 's')  # m2 wird in Grün dargestellt
plt.scatter(*m3, color='blue', marker = 's')  # m3 wird in Blau dargestellt
plt.show()
```

#### 2. berechnen die Abstände jedes Beispiels (x,y) zu jedem Zentroid m.

Dazu verwenden wir die Euklidische Formel. Für ein Zenroidpunkt $m(x_1, y_1)$  und ein Beispiel $(x_2, y_2)$ 
$$D = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 }$$


```python
def euklidische_distanz(df, m):
    return np.sqrt((df['X'] - m[0])**2 + (df['Y'] - m[1])**2)

liste_distanzen = []
for i, m in enumerate([m1,m2, m3]):
    temp = 'Distanz_m'+str(i+1)
    df[temp] = euklidische_distanz(df, m)
    liste_distanzen.append(temp)

print(liste_distanzen)
df
```

#### 3. Die Beispiele werden dem Zentroid (Cluster) zugeordnet, dem sie am nächsten sind.


```python
def clustering(liste_distanzen):
    df['Cluster'] = df[liste_distanzen].idxmin(axis=1)
    print(df['Cluster'])
    # Ändern der Werte in der Spalte 'Cluster'
    df['Cluster'] = df['Cluster'].map({liste_distanzen[0]: 'C1', liste_distanzen[1]: 'C2', liste_distanzen[2]: 'C3'})

clustering(liste_distanzen)
df
```

```python
colors = {'C1':'red', 'C2':'green', 'C3':'blue'}

# Create a scatter plot
plt.figure(figsize=(10,6))

labels = {'C1':'Cluster 1', 'C2':'Cluster 2', 'C3':'Cluster 3'}

for cluster in np.unique(df['Cluster']):
    plt.scatter(df[df['Cluster']==cluster]['X'], df[df['Cluster']==cluster]['Y'], color = colors[cluster], label=cluster)
    
#for i in range(len(df['X'])):
 #   plt.scatter(df['X'][i], df['Y'][i], color = colors[df['Cluster'][i]], label=df['Cluster'][i])

# Plot the centroids
plt.scatter(*m1, marker='s', s=100, linewidths=1, color='black', zorder=10, label='Centroid 1')
plt.scatter(*m2, marker = 's', s=100, linewidths=1, color='black', zorder=10, label='Centroid 2')
plt.scatter(*m3, marker = 's', s=100, linewidths=1, color='black', zorder=10, label='Centroid 3')

plt.title('Clustered Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

#### 4. Neue Zentroiden werden berechnet und in der Mitte des Clusters platziert.
Die Berechnung erfolgt mit einer Durschnittsberechnung von den x- und y-werten der Datenpunkte zugehörig zu jedem Cluster.

```python
def Zentroid_berechnen(cluster):
    return (df[df['Cluster'] == cluster]['X'].mean(), df[df['Cluster'] == cluster]['Y'].mean())
# Berechnung der neuen Zentroiden
clusters = df['Cluster'].unique()
clusters.sort()
print(clusters)

m1_new = Zentroid_berechnen(clusters[0])#(df[df['Cluster'] == 'C1']['X'].mean(), df[df['Cluster'] == 'C1']['Y'].mean())
m2_new = Zentroid_berechnen(clusters[1])#(df[df['Cluster'] == 'C2']['X'].mean(), df[df['Cluster'] == 'C2']['Y'].mean())
m3_new = Zentroid_berechnen(clusters[2])#(df[df['Cluster'] == 'C3']['X'].mean(), df[df['Cluster'] == 'C3']['Y'].mean())

print("Neuer Zentroid m1: ", m1_new)
print("Neuer Zentroid m2: ", m2_new)
print("Neuer Zentroid m3: ", m3_new)
```

```python
colors = {'C1':'red', 'C2':'green', 'C3':'blue'}

# Create a scatter plot
plt.figure(figsize=(10,6))

labels = {'C1':'Cluster 1', 'C2':'Cluster 2', 'C3':'Cluster 3'}

for cluster in np.unique(df['Cluster']):
    plt.scatter(df[df['Cluster']==cluster]['X'], df[df['Cluster']==cluster]['Y'], color = colors[cluster], label=cluster)
    
#for i in range(len(df['X'])):
 #   plt.scatter(df['X'][i], df['Y'][i], color = colors[df['Cluster'][i]], label=df['Cluster'][i])

# Plot the centroids
plt.scatter(*m1_new, marker='s', s=100, linewidths=1, color='black', zorder=10, label='Centroid 1')
plt.scatter(*m2_new, marker = 's', s=100, linewidths=1, color='black', zorder=10, label='Centroid 2')
plt.scatter(*m3_new, marker = 's', s=100, linewidths=1, color='black', zorder=10, label='Centroid 3')

plt.title('Clustered Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

#### 5. Die Schritte 1 bis 4 werden wiederholt, bis es keine weitere Neuzuordnung von Beobachtungen zu Clustern gibt.

```python
liste_distanzen = []
for i, m in enumerate([m1_new,m2_new, m3_new]):
    temp = 'Distanz_m'+str(i+1)
    df[temp] = euklidische_distanz(df, m)
    liste_distanzen.append(temp)

print(liste_distanzen)
df
```

#### Die Spalte `Cluster` hier ist anders als in der ersten Iteration, d.h. wir wiederholen die Schritte weiter bis sich die Spalte `Cluster` nicht mehr ändert.

### Zusammengefasst sieht der Code so aus:

```python
# Define a function to calculate Euclidean distance
def euklidische_distanz(df, m):
    return np.sqrt((df.iloc[:, 0] - m[0])**2 + (df.iloc[:, 1] - m[1])**2)

# Define a function to perform clustering
def clustering(df, centroids):
    print('centroids: ')
    print(centroids)
    liste_distanzen = []
    for i, m in enumerate(centroids):
        temp = 'Distanz_m'+str(i+1)
        df[temp] = euklidische_distanz(df, m)
        liste_distanzen.append(temp)
    df['Cluster'] = df[liste_distanzen].idxmin(axis=1)
    df['Cluster'] = df['Cluster'].map({'Distanz_m'+str(i+1): 'C'+str(i+1) for i in range(len(centroids))})

def K_Mean_clustering(df, centroids):
    # Perform initial clustering
    clustering(df, centroids)
    
    # Repeat clustering until centroids do not change significantly
    while True:
        new_centroids = [tuple(df[df['Cluster'] == 'C'+str(i+1)].iloc[:, :2].mean()) for i in range(len(centroids))]
        if np.allclose(centroids, new_centroids):
            break
        centroids = new_centroids
        clustering(df, centroids)
        print(df)
    return df

# Define your data
data = {
    'X': [2, 2, 5, 4, 8, 6, 5, 5, 6, 4],
    'Y': [4, 6, 6, 7, 3, 6, 2, 7, 3, 4]
}
df = pd.DataFrame(data)

# Define k centroids
centroids = [(1, 5), (4, 1), (8, 4)]  # Add or remove centroids as needed

K_Mean_clustering(df, centroids)
```

### Leichter geht es so:

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def kompakte_K_Mean_clustering(df, n_clusters):
    X = df.values

    clust = KMeans(n_clusters=n_clusters, n_init=10).fit(X)
    y = clust.predict(X)

    # Assign the cluster labels to each data point
    df['Cluster'] = clust.labels_

    # Plot the data points with different colors for each cluster
    colors = ['red', 'green', 'blue']  # Add more colors if you have more than 3 clusters

    plt.figure(figsize=(10,6))
    for i in range(n_clusters):
        plt.scatter(df[df['Cluster']==i].iloc[:, 0], df[df['Cluster']==i].iloc[:, 1], color = colors[i])

    # Get the coordinates of the cluster centers
    centroids = clust.cluster_centers_
    print('Zentroiden: \n',centroids)

    #Verzerrungswert
    print('Verzerrungswert: \n',clust.inertia_)

    # Plot the centroids
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='s', s=100, linewidths=1, color='black', zorder=10)
    plt.title('Clustered Data')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()

    return df

# Erstellen Sie den DataFrame
E_1 = pd.DataFrame({
    'x1': [2, 4, 2.8, 1.9, 4.1, 3.5, 3.8, 2, 2.9, 4],
    'x2': [3, 4.2, 2.1, 2, 3.2, 3.5, 4.1, 1.9, 3.1, 3.9]
})

# Ausgabe des DataFrames
print(E_1)

# Call the function
clustered_df = kompakte_K_Mean_clustering(E_1, 2)
print(clustered_df)
```
